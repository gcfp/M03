# CORRELACIÃ“N

```python
import seaborn as sns
import pandas as pd
import numpy as np

df_iris = pd.read_csv('iris.csv')

sns.scatterplot(data = df_iris, x='petal.length', y='petal.width')
sns.lmplot(data = df_iris, x='petal.length', y='petal.width')
```

```python
# Matriz de correlacion
df_iris.corr()
np.corrcoef(df_iris['petal.length'], df_iris['petal.width'])[0,1]
```

```python
sns.heatmap(df_iris.corr(), annot=True, cmap='coolwarm')
```

## Ejercicio 2

Existe relaciÃ³n entre el nÃºmero de habitantes y el % de voto a obama?


```python
# MuÃ©stralo grÃ¡ficamente mediante un scatter-plot (escala el total de votos /1000)
df_obama = pd.read_csv('2008_all_states.csv')
df_obama['total_votes_k'] = df_obama['total_votes']/1000
sns.lmplot(data = df_obama, x='total_votes_k', y='dem_share')

# Obten la correlaciÃ³n
np.corrcoef(df_obama.total_votes_k, df_obama.dem_share)[0,1]
```
