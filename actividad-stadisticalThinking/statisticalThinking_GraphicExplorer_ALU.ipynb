### HISTOGRAMA


Vamos a importar el dataset Iris un dataset generado por el botànico Edward Anderson donde analizaba tres espècies diferentes de la planta Iris y que da mucho juego a nivel estadìstico

```python
#importamos dataset
import pandas as pd

df_iris = pd.read_csv('iris.csv')
```

```python
# ojeamos dataset
df_iris.head()
```

```python
# creamos una serie solo con la longitud de un tipo en conreto
versicolor_petal_lenght = df_iris[df_iris['variety'] == 'Versicolor']['petal.length']
setosa_petal_lenght = df_iris[df_iris['variety'] == 'Setosa']['petal.length']
virginica_petal_lenght = df_iris[df_iris['variety'] == 'Virginica']['petal.length']
versicolor_petal_lenght
```

```python
# import plotting modules
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


# Le añado el estilo seaborn por defecto

sns.set()

#num Bins aconsejable

nbins = int(np.sqrt(len(versicolor_petal_lenght)))

# Muestro histograma

plt.hist(versicolor_petal_lenght, bins=20)


#Añado etiquetas a ejes
plt.xlabel('longitud petalo cm')
plt.ylabel('Numero total')

plt.show()
```

### BEE SWARM PLOT (PANEL DE ABEJAS)

```python
#creamos panel de abejas
sns.swarmplot(x='variety', y='petal.length', data = df_iris)
plt.xlabel('variedad')
plt.ylabel('longitud petalo cm')

plt.show()
```

```python
np.std(versicolor_petal_lenght)
```

```python
np.std(virginica_petal_lenght)
```

### ECDF

estimador de funcion de distribucion acumulada

```python
def ecdf(data):
  #Numero de datos
  n = len(data)

  #ordenamos datos
  x = np.sort(data)

  # valores y
  y = np.arange(1,n+1) / n

  return x, y
```

```python
x_vers, y_vers = ecdf(versicolor_petal_lenght)
plt.plot(x_vers, y_vers, marker='.')

plt.xlabel('longitud de petalo')
plt.ylabel('ECDF')

plt.margins(0.02)

plt.show()
```
